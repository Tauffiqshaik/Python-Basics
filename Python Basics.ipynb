# Nested if Statement


```python
#Nested if conditions means if inside if
attendance=int(input("Enter the attendance value"))
if attendance>=75:
    marks=float(input("Enter the marks"))
    if marks>=35:
        print("pass")
    else:
        print("Fail")
else:
    print("Not Eligible to write Exams")
    
```
